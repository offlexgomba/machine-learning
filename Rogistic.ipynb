```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(42)

x = np.random.rand(100) * 100  
noise = np.random.randn(100) * 5  
y = 2 * x + 3 + noise

data = pd.DataFrame({'House Size': x, 'House Price': y})

stats = data.describe()

print("Basic Statistics:\n", stats)

plt.scatter(x, y, label="Data points", color="black", alpha=0.6)
plt.xlabel("House Size")
plt.ylabel("House Price")
plt.title("Synthetic Dataset for Linear Regression")
plt.legend()
plt.show()
```

```python
def split_data(x, y, train_ratio=0.8, seed=42):
    np.random.seed(seed) 
    indices = np.random.permutation(len(x))  
    
    train_size = int(len(x) * train_ratio)  
    train_indices = indices[:train_size]  
    test_indices = indices[train_size:]  
    
    x_train, y_train = x[train_indices], y[train_indices]
    x_test, y_test = x[test_indices], y[test_indices]
    
    return x_train, x_test, y_train, y_test

np.random.seed(42)
x = np.random.rand(100) * 100  
noise = np.random.randn(100) * 5  
y = 2 * x + 3 + noise  

x_train, x_test, y_train, y_test = split_data(x, y)

print(f"Training set size: {len(x_train)}\nTesting set size: {len(x_test)}")
```
